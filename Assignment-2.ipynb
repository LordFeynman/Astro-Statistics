## Question 1: Probability distribution of the sum and differences of the values

Let's start considering two dices:
- Die 1: $\{1,2,3,4,5,6\}$
- Die 2: $\{1,2,3,4,5,6\}$

From this configuration we can already see that the lowest sum possible is $2$ and the highest sum is $12$. The outcomes in between would be $\{3,4,5, \dotso,11,12\}$ Therefore, the possible outcomes are:

The probability distribution is then,
\begin{equation}
P(x)=\left\{\begin{array}{rc}

\frac{8}{36} \quad , &\mbox{if} \quad x  = 2\\
\frac{4}{36} \quad , &\mbox{if} \quad x  = 4\\
\frac{2}{36} \quad , &\mbox{if} \quad x  = 5\\

\end{equation}

The Python code below provides a graphical view of the dice throw as well as the probabilities. We will modify this code to adapt it to a more generic situation, but we decided to leave it as simple as possible at this point for the sake of clarity.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

n = 100000
r = np.linspace(2,12,13)
count = np.zeros(13)
out = []

def roll_dice(n = 100000):
    
    for i in range(n):
        # Roll each dice
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        
        # Sum the values to get the score
        score = die1 + die2
        out.append(score)
        
        # Conditions to add to the count
        for j in range(len(r)):
            if score == j:
                count[j] += 1
    return count/n

probability = roll_dice()
print("Probabilities are: ", np.round(probability*100,2),"%")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
fig.suptitle('Probability Distribution')
ax1.scatter(r, probability)
ax1.set_title('Scatter Plot')
ax1.set(xlabel='$X$', ylabel='$P(X)$')
ax2.hist(out, bins = np.arange(2, 14), edgecolor='black')
ax2.set_title('Histrogram')
ax2.set(xlabel='Outcome', ylabel='Frequency')

fig.tight_layout()
plt.show()
```

## Question 2: Probability distribution for 100 dice

First of all, let's calculate the mean and standard deviation of one die. The straightforward approach is to use the know equations, which gives us the mean
\begin{eqnarray}
\mu &=& \frac{\sum_n x_i}{n} = 3.5
\end{eqnarray}
and the standard deviation
\begin{eqnarray}
\sigma^2 &=& \frac{\sum_n (x_i - \bar{x})^2}{n} = \frac{35}{12}
\end{eqnarray}

Therefore, we can use the Central-Limit Theorem to estimate the mean and variance of 100 dice throws, which will be
\begin{equation}
\mu_{100} = 350
\end{equation}

\begin{equation}
\sigma^2_{100} = \frac{3500}{12} \approx 291
\end{equation}

The Python code down below is used to plot the graph of the distribution. Note that this is a more general code; we could simply change the variable `n_dice` to $2$ and we would obtain the results in Q1. Another interesting thing to notice is how similar this distribution is to a Gaussian, which provides further evidence for the Central-Limit Theorem.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10000
n_dice = 100                           # Number of dice used
r = np.linspace(0,6*n_dice,6*n_dice+1)
die = np.zeros(n_dice)
count = np.zeros(6*n_dice+1)
#out = []

def roll_dice(n = 10000):
    start = 1
    while start < n:
        # Roll each dice
        roll = np.random.randint(1,7,n_dice)
        
        # Sum the values to get the score
        for i in range(n_dice):
            score = sum(roll)
#            out.append(score)
        
        # Conditions to add to the count
        for k in range(len(r)):
            if score == k:
                count[k] += 1
        start += 1
    return count/n

probability = roll_dice()
#print(probability)

#plt.figure(1, dpi = 120)
#plt.hist(out, edgecolor='black')
#plt.show()

plt.figure(1, dpi = 120)
plt.scatter(r, probability, s=5)
plt.xlabel("$X$")
plt.ylabel("$P(X)$")
plt.title("Probability distribution for the sum of 100 dice")
plt.show()
```

## Question 3: How to label two dice in order to have a uniform distribution

In this problem we want to find a way to relabel two six-sided dice so that all integers $x$ with $1 \leqslant x \leqslant 12$ are equally likely to appear as the sum of the dice. We can propose to label the dice as
- Die 1: $\{0,0,0,6,6,6\}$
- Die 2: $\{1,2,3,4,5,6\}$

We can see the possible outcomes in the Table below:

\begin{array}{c|cccccc}
 & \textbf{0} & \textbf{0} & \textbf{0} & \textbf{6} & \textbf{6} & \textbf{6} \\
\hline
\textbf{1} & 1 & 1 & 1 & 7 & 7 & 7 \\
\textbf{2} & 2 & 2 & 2 & 8 & 8 & 8 \\
\textbf{3} & 3 & 3 & 3 & 9 & 9 & 9 \\
\textbf{4} & 4 & 4 & 4 & 10 & 10 & 10 \\
\textbf{5} & 5 & 5 & 5 & 11 & 11 & 11 \\
\textbf{6} & 6 & 6 & 6 & 12 & 12 & 12 \\
\end{array}

We can see that the sums go from 1 to 12 and each sum appears thrice. This already tells us that the probability is uniform and equals to $\frac{3}{36}$. One question that might arise is "Why did we choose this particular combination?"; well, looking at the table above we can see that any other combination would result in two things:
- We would be out of the range $1 \leqslant x \leqslant 12$, for instance, one die $\{1,1,1,6,6,6\}$ would exclude the sum 1 from the outcomes.
- We would have a spike in one particular sum. If we take the die $\{1,1,1,6,6,6\}$ again we would have twice the amount of 7s, giving every other sum a probability $\approx 0.8$ and a probability of $\approx 0.16$ for the 7.

Everytime we change the values of the first half of our die we change the range of the outcomes and increase the number of spikes. So, for example, the die $\{2,2,2,6,6,6\}$ produces outcomes in the range $3 \leqslant x \leqslant 12$ and favor the results 7 and 8, while the die $\{6,6,6,6,6,6\}$ has uniform distribution but is in the range $6 \leqslant x \leqslant 12$.

Formally, we can call $X$ the values of the first die and $Y$ the values of the second. Then, we want to evaluate $P(X+Y = k)$, i.e. the probability that the values $X$ and $Y$ sum up to $k$. This leads us to
\begin{equation}
P(X+Y = k) = \frac{1}{2}\cdot\frac{1}{6} = \frac{1}{12} \quad \hbox{for} \quad 1 \leqslant k \leqslant 6,
\end{equation}

when we need to roll a $0$ on the first die and $k$ on the second, and

\begin{equation}
P(X+Y = k) = \frac{1}{2}\cdot\frac{1}{6} = \frac{1}{12} \quad \hbox{for} \quad 7 \leqslant k \leqslant 12,
\end{equation}

when we need to roll a $6$ on the first die and $k$ on the second.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

n = 100000
r = np.linspace(2,12,11)
count = np.zeros(11)
out = []

def roll_dice(n = 100000):
    
    for i in range(n):
        # Roll each dice
        die1 = random.choice([0,0,0,6,6,6]) # Now our first die is modified
        die2 = random.randint(1, 6)
        
        # Sum the values to get the score
        score = die1 + die2
        out.append(score)
        
        # Conditions to add to the count
        for j in range(len(r)):
            if score == j+2:
                count[j] += 1
    return count/n

probability = roll_dice()
print("Probabilities are: ", np.round(probability*100,2),"%")

plt.figure(1, dpi = 120)
plt.hist(out, bins = np.arange(1, 14), edgecolor='black')
plt.xlabel("Outcomes")
plt.ylabel("Frequency")
plt.title("Distribution for relabeled dice $\{0,0,0,6,6,6\}$ and $\{1,2,3,4,5,6\}$ ")
plt.show()
```

## Question 4: 100 dice with uniform distribution

This time we want to find a way to relabel 100 six-sided dice with integers so that all the values are equally likely to appear as the sum of the dice. Note that this time it is not required to play all values in a certain range and not even in what range. Thus, we can suggest to label the $r^{th}$ die in this way:
- Die r: $\{0,1,2,3,4,5\} \times 10^r$ with $r \in [0,100[$

With 100 six-sided dice, the total amount of possible results we can have is:

$$ \hbox{Combinations} = 6^{100} \sim 6,5 \cdot 10^{77}$$

Furthermore, every value for the sum of the 100 dice can be obtained with just one combination. Therefore, we can obtain the probability of an outcome $X$ with
\begin{equation}
P(X)=\frac{1}{6^{100}} \sim 1,5 \cdot 10^{-78}
\end{equation}

which is very small, meaning that we will need a lot of simulation to reproduce all the values.

It is worth noticing that we can use any power with a base greater than $5$. ## Question 6: Distribution plots of Q4 for $N = 2, 3, 4, 10$

If we change the number of dice the values of the sum for each throw follow an uniform distribution and not a gaussian one. In this case each die is labelled in a different way, thus we cannot apply the central limit theorem because each random variable follows a different distribution. The following programm shows the uniform distribution we expect.

**Disclaimer:** The programm below is a general one. Unfortunately, the number of iterations and the number of dice used leads to a computational issue: time. One can choose to use fewer iterations, but in that case the result is not as clear as we would like. In this code we only used 2 dice, since we don't have computer power to increase the simulation. Nonetheless, the result is satisfactory and can be extrapolated for higher numbers.

```python
import numpy as np
import matplotlib.pyplot as plt

# define the variables we are going to use

n = 1000000   # it will take some time but with less iteration the result is not as clear
n_dice = 2 #Enjoy changing the number of dice
k = np.zeros(n_dice)
out = []

for i in range(n_dice):
    k[i] = 10**i
    
for i in range(n):
    # roll all the dice
    roll = np.random.choice([0,1,2,3,4,5],n_dice)*k
    # sum the values to get the score
    score = sum(roll)
    out.append(score)

max_value=max(out)
plt.figure(1, dpi = 120)
plt.hist(out, bins = np.arange(0, max_value+2), edgecolor='black')
plt.xlabel("Outcomes")
plt.ylabel("Frequency")
plt.title("Distribution for N relabeled dice ")
plt.show()
```
